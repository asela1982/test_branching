```python
# load csv file and pull list of zipcodes

filepath = os.path.join("df_city.csv")
city_df = pd.read_csv(filepath)
zipcodes = city_df
zipcodes = zipcodes['zip']
```

```python
len(zipcodes)
```

```python
# build loop to create csv file out of venue data

city = []
lat = []
lon = []
name = []
rating = []
rating_count = []
state = []

counter = 0

for zip in zipcodes:
    params = {
        "key": meetup_key,
        "text": zip,
        "page": 200
    }
    
    request = requests.get(url, params=params)
    data = request.json()
    len_call = len(data)
    
    print(f"{counter} {len_call}")
    counter = counter + 1
    
    if int(request.headers['X-RateLimit-Remaining']) <= 2:
        time.sleep(60)
    
    for i in range(len_call):
        city.append(data[i]['city'])
        lat.append(data[i]['lat'])
        lon.append(data[i]['lon'])
        name.append(data[i]['name'])
        rating.append(data[i]['rating'])
        rating_count.append(data[i]['rating_count'])

```

```python
data = pd.DataFrame({"city": city, "lat": lat, "lon": lon, "name": name, "rating": rating, "rating_count": rating_count})
data
```
