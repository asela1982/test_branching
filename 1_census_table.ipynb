```python
# import dependencies
import pandas as pd
import os
import csv
```

### load the datasets to pandas dataframe

```python
#load the "2017_Gaz_zcta_national.txt" to a a pandas dataframe
result = {}

filepath = os.path.join("data_input","2017_Gaz_zcta_national.txt")

with open(filepath,'r') as f:
    reader = csv.DictReader(f,delimiter='\t')
    for row in reader:
        for column, value in row.items():
            result.setdefault(column, []).append(value)
            
df_zcta = pd.DataFrame.from_dict(result)
display(df_zcta.head())
display(df_zcta.shape)
```

```python
#load the "ACS_15_5YR_B01002_with_ann.csv"(MEDIAN AGE BY SEX) to a a pandas dataframe
result = {}

filepath = os.path.join("data_input","ACS_15_5YR_B01002","ACS_15_5YR_B01002_with_ann.csv")

with open(filepath,'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        for column, value in row.items():
            result.setdefault(column, []).append(value)
  
df_age = pd.DataFrame.from_dict(result)
df_age.columns = df_age.iloc[0,:].values
df_age.drop(df_age.index[0],inplace=True)
display(df_age.head())
display(df_age.shape)
```

```python
#load the "ACS_15_5YR_B02001_with_ann"(RACE) to a a pandas dataframe
result = {}

filepath = os.path.join("data_input","ACS_15_5YR_B02001","ACS_15_5YR_B02001_with_ann.csv")

with open(filepath,'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        for column, value in row.items():
            result.setdefault(column, []).append(value)
            
df_race = pd.DataFrame.from_dict(result)
df_race.columns = df_race.iloc[0,:].values
df_race.drop(df_race.index[0],inplace=True)
display(df_race.head())
display(df_race.shape)
```

```python
#load the "ACS_15_5YR_B19013_with_ann"(MEDIAN HOUSEHOLD INCOME) to a a pandas dataframe
result = {}

filepath = os.path.join("data_input","ACS_15_5YR_B19013","ACS_15_5YR_B19013_with_ann.csv")

with open(filepath,'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        for column, value in row.items():
            result.setdefault(column, []).append(value)
            
df_income = pd.DataFrame.from_dict(result)
df_income.columns = df_income.iloc[0,:].values
df_income.drop(df_income.index[0],inplace=True)
display(df_income.head())
display(df_income.shape)
```

```python
#load the "ACS_15_5YR_S1501_with_ann"(EDUCATIONAL ATTAINMENT) to a a pandas dataframe
result = {}

filepath = os.path.join("data_input","ACS_15_5YR_S1501","ACS_15_5YR_S1501_with_ann.csv")

with open(filepath,'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        for column, value in row.items():
            result.setdefault(column, []).append(value)
            
df_edu = pd.DataFrame.from_dict(result)
df_edu.columns = df_edu.iloc[0,:].values
df_edu.drop(df_edu.index[0],inplace=True)
display(df_edu.head())
display(df_edu.shape)
```

```python
#load the "ACS_15_5YR_B01003_with_ann"(population) to a a pandas dataframe
result = {}

filepath = os.path.join("data_input","ACS_15_5YR_B01003","ACS_15_5YR_B01003_with_ann.csv")

with open(filepath,'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        for column, value in row.items():
            result.setdefault(column, []).append(value)
            
df_pop = pd.DataFrame.from_dict(result)
df_pop.columns = df_pop.iloc[0,:].values
df_pop.drop(df_pop.index[0],inplace=True)
display(df_pop.head())
display(df_pop.shape)
```

### Feature engineering

```python
# obtain the weighted education score
df_edu['edu_score']= ''
for index, row in df_edu.iterrows():
    temp = int(row[4])*1+int(row[5])*2+int(row[6])*3+int(row[7])*4+int(row[8])*5+int(row[9])*6+int(row[10])*7
    try:
        weight = int(temp)/int(row[3])
        df_edu.set_value(index,'edu_score',weight)
    except ZeroDivisionError:
        df_edu.set_value(index,'edu_score',None)       
```

```python
df_edu.head()
```

```python
# obtain the diversity index
df_race['div_score'] = ''
for index, row in df_race.iterrows():
    try:
        other_races = (int(row[3])-int(row[4]))/int(row[3])
        df_race.set_value(index,'div_score',other_races)   
    
    except ZeroDivisionError:
        df_race.set_value(index,'div_score',None) 
```

```python
df_race.head()
```

### Construct the census summary table

```python
# merge age
df_census = df_zcta.merge(df_age[['Id2','Estimate; Median age -- - Total:']],left_on='GEOID',right_on='Id2',how='inner')
df_census.drop('Id2', axis=1, inplace=True)
df_census.head()
```

```python
# merge income
df_census = df_census.merge(df_income[['Id2','Estimate; Median household income in the past 12 months (in 2015 Inflation-adjusted dollars)']],left_on='GEOID',right_on='Id2',how='inner')
df_census.drop('Id2', axis=1, inplace=True)
df_census.head()
```

```python
# merge edu score
df_census = df_census.merge(df_edu[['Id2','edu_score']],left_on='GEOID',right_on='Id2',how='inner')
df_census.drop('Id2', axis=1, inplace=True)
df_census.head()
```

```python
# merge diversity score
df_census = df_census.merge(df_race[['Id2','div_score']],left_on='GEOID',right_on='Id2',how='inner')
df_census.drop('Id2', axis=1, inplace=True)
df_census.head()
```

```python
# merge population
df_census = df_census.merge(df_pop[['Id2','Estimate; Total']],left_on='GEOID',right_on='Id2',how='inner')
df_census.drop('Id2', axis=1, inplace=True)
df_census.head()
```

```python
#rename column names
df_census.columns = ['ALAND', 'ALAND_SQMI', 'AWATER', 'AWATER_SQMI', 'GEOID', 'INTPTLAT',
       'INTPTLONG','Median_age', 'Median_household_income', 'Educational_score',
       'Diversity_score','population']
```

```python
# export the census data to data_output
df_census.to_csv("data_output/df_census.csv", encoding="utf-8", index=False)
```
